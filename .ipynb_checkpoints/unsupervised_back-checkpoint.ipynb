```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle 
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import davies_bouldin_score, silhouette_score
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv("Bank_of_America_data.csv")
data_bl = data.dropna()

features_bl = data_bl.iloc[:,1:]
target_bl = data_bl.iloc[:,0]
```

```python
ridge_gs = GridSearchCV(Ridge(),{'alpha' : np.logspace(-5,5,40)}, cv=10)

encoder = LabelEncoder()
one_hot_encoder = OneHotEncoder()
standard = StandardScaler()
```

```python
ridge_gs.fit(data_bl.MORTDUE.values.reshape(-1,1),data_bl.VALUE)
lm = ridge_gs.best_estimator_
```

```python
data_fe = data.dropna(thresh=5)
data_fe.REASON.fillna("other",inplace=True)
data_fe.JOB.fillna("Other",inplace=True)
data_fe.DEROG.fillna(0,inplace=True)
data_fe.YOJ.fillna(0,inplace=True)

#data_fe.CLNO.fillna(0,inplace=True)
data_fe.NINQ.fillna(0,inplace=True)
data_fe.DELINQ.fillna(0,inplace=True)

#adding new column for debinc
#data_fe = pd.concat([data_fe, pd.Series(~data_fe.DEBTINC.isna()*1,index=data_fe.index,name="debtincisnull")],axis=1)
data_fe.CLAGE.fillna(data.CLAGE.mean(),inplace=True)
data_fe.DEBTINC.fillna(data.DEBTINC.mean(),inplace=True)
data_fe = data_fe[(~data_fe.MORTDUE.isna())|(~data_fe.VALUE.isna())]

data_fe.MORTDUE.fillna(data_fe.VALUE/lm.coef_ - lm.intercept_,inplace=True)
data_fe.VALUE.fillna(data_fe.MORTDUE*lm.coef_ + lm.intercept_, inplace=True)
```

```python
data_fe=data_fe.dropna()
features_fe=data_fe.iloc[:,1:]
target_fe=data_fe.iloc[:,0]
```

```python
features_fe_numeric = pd.DataFrame(standard.fit_transform(features_fe.drop(['JOB','REASON'],axis=1)), columns=features_fe.columns.drop(['JOB','REASON']), index=features_fe.index)
```

```python
pca = PCA(n_components=2)
```

```python
features_fe_numeric = pca.fit_transform(features_fe_numeric)
```

```python
pca.explained_variance_ratio_
```

```python
plt.figure(figsize=(20,10))
sns.scatterplot(features_fe_numeric[:,0],features_fe_numeric[:,1],hue=target_fe)

plt.plot([-15,15],[0,0],ls="--")
plt.plot([0,0],[-10,10],ls="--")
plt.title("PCA")
plt.xlim([-10,10])
plt.ylim([-10,10])
plt.xlabel("Composant 1 : %s "%round(100*pca.explained_variance_ratio_[0],2))
plt.ylabel("Composant 1 : %s "%round(100*pca.explained_variance_ratio_[1],2))
plt.savefig("pca.png")
```

```python
plt.figure(figsize=(10,10))
for i in range(len(pca.components_[0])):
    plt.plot([0,pca.components_[0][i]],[0,pca.components_[1][i]])
    plt.text(pca.components_[0][i],pca.components_[1][i],features_fe.columns.drop(['JOB','REASON'])[i], fontsize=14)
plt.xlim([-0.8,0.8])
plt.ylim([-0.8,0.8] )
plt.plot([-1,1],[0,0],ls="--")
plt.plot([0,0],[-1,1],ls="--")
circle1=plt.Circle((0,0),0.8,color='k', fill=False)
plt.gcf().gca().add_artist(circle1)
plt.savefig("pca_comp.jpg")
```

```python
jobs = pd.DataFrame(one_hot_encoder.fit_transform(features_bl.JOB.to_numpy().reshape(-1,1)).toarray(),index=features_bl.index).drop(5,axis=1)
features_bl_numeric = pd.concat([features_bl.drop(['JOB'],axis=1),jobs],axis=1) 
features_bl_numeric.REASON = encoder.fit_transform(features_bl_numeric.REASON)
```

```python
features_bl_numeric = pd.DataFrame(standard.fit_transform(features_bl_numeric), columns=features_bl_numeric.columns, index=features_bl_numeric.index)
```

```python
distortions = []
silhouttes = []
for k in range(2,11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(features_bl_numeric)
    distortions.append(davies_bouldin_score(features_bl_numeric, kmeans.predict(features_bl_numeric)))
    silhouttes.append(silhouette_score(features_bl_numeric, kmeans.predict(features_bl_numeric)))
```

```python
plt.figure(figsize=(16,8))
plt.plot(range(2,11), distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('intra / inter ')
plt.title('intra /inter inertie showing the optimal k')
plt.savefig("clustering_intra_score.png")
```

```python
fig = plt.figure(figsize=(16,8))
plt.plot(range(2,11), silhouttes, 'bx-')
plt.xlabel('k')
plt.ylabel('intra / inter ')
plt.title('silhouette method showing the optimal k')
plt.savefig("clustering_silhouette_score.png")
```
